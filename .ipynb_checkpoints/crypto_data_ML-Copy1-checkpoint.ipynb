```python
# Import Dependencies
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Read crypto_data.csv
df=pd.read_csv('crypto_data.csv')

# Display the 5 first rows of the data frame
df.head(10)
```

```python
# Display the info of the data frame
df.info()
```

```python
# Filter the data frame for cryptocurrencies being traded only
df_trading=df[df['IsTrading']==True]

# Display the info of the data frame
df_trading.info()
```

```python
# Drop 'IsTrading' column
df_trading=df_trading.drop('IsTrading',axis=1)

# Display the 5 first rows of the data frame
df_trading.head()
```

```python
# Remove rows with at least one null value
df_cleaned=df_trading.dropna(axis=0)

# Display the info of the data frame
df_cleaned.info()
```

```python
# Filter the data frame for cryptocurrencies that have been mined
df_mined=df_cleaned[df_cleaned['TotalCoinsMined']>0]

# Display the info of the data frame
df_mined.info()
```

```python
# Drop the coin name columns
df_final=df_mined.drop(['CoinName','Unnamed: 0'],axis=1)

# Convert the data type of the 'TotalCoinSupply' column
df_final['TotalCoinSupply'] = df_final['TotalCoinSupply'].astype(float)

# Display the info of the data frame
df_final.info()
```

```python
# Convert categorical data into numerical data
df_dummies=pd.get_dummies(df_final,columns=['Algorithm','ProofType']).reset_index(drop=True)

# Print the columns
print(df_dummies.columns)

# Display the data frame
df_dummies
```

```python
# Standardize the data using StandardScaler
crypto_scaled=StandardScaler().fit_transform(df_dummies)
```

```python
# Initialize PCA model
pca = PCA(n_components=0.90)

# Get 90% principal components for the crypto data.
pc_pca = pca.fit_transform(crypto_scaled)

# Check for the sum of the explained variance ration
print(sum(pca.explained_variance_ratio_))

# Display the shape of the PCA model output
pc_pca.shape
```

## PCA has reduced the number of features from 98 to 74 keeping a %90 explained variance ratio.

```python
# Run t-SNE
tsne = TSNE(learning_rate=250)
tsne_features = tsne.fit_transform(pc_pca)
```

```python
# plot the t-SNE output
df_tsne=pd.DataFrame(tsne_features,columns=['x','y'])
df_tsne.plot(kind='scatter',x='x',y='y')
plt.show()
```

```python
# Finding the best value for k
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(df_tsne)
    inertia.append(km.inertia_)

# Create a data frame to hold the data for the elbow curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.head()
```

```python
# Plot the elbow curve to find the best candidate(s) for k
plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow curve for crypto data')
plt.show()
```

```python
# Initialize the K-Means model and visualize the clusters
model = KMeans(n_clusters=4, random_state=0)
model.fit(df_tsne)
predictions = model.predict(df_tsne)
df_tsne.plot(kind='scatter',x='x',y='y',colormap='viridis',c=predictions, )
plt.show()
```

## According to the above elbow curve, the elbow happens where k=4. Therefore, we can determine that the cryptocurrencies can be clustered in 4 different clusters.
